```python
import os
from PIL import Image
import glob
import numpy as np
```

```python
files = glob.glob("*_search_images_*.png")
images = None
for file in sorted(files):
    im = np.array(Image.open(file).convert('RGB'))[120:220, :, :]
    if images is None:
        images = im
    else:
        images = np.append(images, im, axis=0)
images.shape
```

```python
img = Image.fromarray(images)
img
```

```python
img.save("cifar10_search_images.png")
```
